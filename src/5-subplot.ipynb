### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [15, 30, 45, 10]

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line chart in the first subplot
axes[0, 0].plot(x, y, color='blue')
axes[0, 0].set_title('Line Chart')
axes[0, 0].set_xlabel('X-axis')
axes[0, 0].set_ylabel('Y-axis')

# Bar chart in the second subplot
axes[0, 1].bar(categories, values, color='orange')
axes[0, 1].set_title('Bar Chart')
axes[0, 1].set_xlabel('Categories')
axes[0, 1].set_ylabel('Values')

# Scatter plot in the third subplot
axes[1, 0].scatter(x, y, color='green', marker='o')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('X-axis')
axes[1, 0].set_ylabel('Y-axis')

# Pie chart in the fourth subplot
axes[1, 1].pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=['red', 'yellow', 'blue', 'green'])
axes[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()
plt.show()
```
